# Revenue vs Order Method Type with Quantity

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Order method type', y='Revenue', size='Quantity', sizes=(20, 200), hue='Order method type', data=df)
# Customize the plot
plt.title('Revenue vs Order Method Type with Quantity')
plt.xlabel('Order Method Type')
plt.ylabel('Revenue')
plt.legend(title='Order Method Type')
plt.show()






```

# total quantity and average revenue per order method type

```python
# Calculate total quantity and average revenue per order method type
grouped_data = df.groupby('Order method type').agg({'Revenue': 'mean', 'Quantity': 'sum'}).reset_index()

# Create the plot
fig, ax1 = plt.subplots(figsize=(10, 6))

 # Bar plot for total quantity
sns.barplot(x='Order method type', y='Quantity', data=grouped_data, ax=ax1, color='lightblue')
ax1.set_ylabel('Total Quantity')
ax1.set_title('Total Quantity and Average Revenue per Order Method Type')


 # Line plot for average revenue
ax2 = ax1.twinx()
sns.lineplot(x='Order method type', y='Revenue', data=grouped_data, ax=ax2, marker='o', color='red')
ax2.set_ylabel('Average Revenue')

plt.show()
```

```python
country = df.groupby(["Retailer country"], as_index=False)['Revenue'].sum().sort_values(by='Revenue', ascending=False).head(10)
plt.figure(figsize=(13,5))
sns.barplot(data=country,x="Retailer country",y='Revenue')
plt.title("top 10 countries")
plt.show()
```

```python
plt.figure(figsize=(13,5))
am= sns.barplot(data=country,x="Retailer country",y='Revenue')
for bars in am.containers:
  am.bar_label(bars)
plt.title("top 10 countries")
plt.show()
```
